### Step 4. Discover what is the mean age per occupation

```python
occ_grp=users.groupby(['occupation'])
```

```python
occ_grp['age'].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
filt=users['gender']=='M'
male_occ=users[filt]
male_occ
```

```python
male_table=pd.DataFrame(male_occ['occupation'].value_counts())
male_table
```

```python
filt=users['gender']=='F'
female_occ=users[filt]
female_occ
```

```python
female_table=pd.DataFrame(female_occ['occupation'].value_counts())
female_table
```

```python
male_table['females']=female_table['count']
aggr=male_table
aggr
aggr.rename(columns={'count': 'males'}, inplace=True)
aggr
```

```python
aggr['ratio']=aggr['males']/aggr['females']
aggr
```

```python
aggr.sort_values(by='ratio', ascending=False)
```
